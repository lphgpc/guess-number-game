```python
# Guessing Game Challenge Let's use while loops to create a guessing game.

#The Challenge:

#Write a program that picks a random integer from 1 to 100, 
#and has players guess the number. The rules are:

#If a player's guess is less than 1 or greater than 100, say "OUT OF BOUNDS"
#On a player's first turn, if their guess is
#within 10 of the number, return "WARM!"
#further than 10 away from the number, return "COLD!"
#On all subsequent turns, if a guess is
#closer to the number than the previous guess return "WARMER!"
#farther from the number than the previous guess, return "COLDER!"
#When the player's guess equals the number, tell them they've guessed correctly and how many guesses it took!
#You can try this from scratch, or follow the steps outlined below. 


```

```python
# importer toute la bibiotheque RANDOM
from random import *


# choisir un nombre entre 1 et 100
solu = randint(0,100)

# créer la liste des reponsse. elle seront rajouté au fur et à mesure
guess=[]

# initialisation des variable
conteur = 0
entree = 0

print("cherchez un nombre entre 0 et 100")
print()
print()


# tant que la réponsse est diferent du nombre à trouvé
while entree != solu:
    conteur=conteur+1 
    print("le nombre de tour est: {}".format(conteur))  
    entree = input("entree une valeur entre 0 et 100 :")
    entree = int(entree)
    guess.append(entree)   
    
    # l'ecar est la diference entre le dernier nombre de la liste des entree - la solution
    ecar= guess[conteur-1]-solu
    
    # on prend la valeur absolu de l'ecar
    ecar =abs(ecar)
    
       
    # si on est au deuxieme tour et plus, on peut savoir si on chauf ou on refroidi en comparant l'ecar de la reponsse
    # actuelle ( la derniere [-1]) avec l'ecar de la reponsse precedente(l'avan derniere [-2]).
    if len(guess)>1:
        if abs(guess[-1]-solu)< abs(guess[-2]-solu):
            print("WARMER")
        else:
            print("COLDER")
    else:
         # si l'ecart et moins de 10 ou plus de 10
        if (ecar) >= 10:
            print ("COLD")
        else:
            print ("WARM")
            
# si on trouve la reponsse  alors fin.
print()
print()

print ("Voila la liste de vos reponsses {}".format(guess))
print("BRAVO, la reponsse est {} en effet. you found it in {} coup !".format(guess[-1],conteur))
```
